##Explore data about Hurricane Isaac power outages

```python
from pandas import *
import matplotlib.pyplot as plt
figsize(8,6)
pandas.__version__
```

Import excel data  
Use 'parse_dates' to tell it to combine column 4 (Date) and column 5 (Time) as one Date-Time column  
Assign the Date-Time column as the index as well

```python
xls = ExcelFile('OutageData_MasterList.xlsx')
xls.use_xlsx
outage_data = xls.parse('OutagesAll', parse_dates = [[4,5]], index_col=0) 
outage_data
```

Compute elapsed time from DateTimeIndex and append a new column with the computed values

```python
outage_data['Elapsed DHMS'] = outage_data.index.to_pydatetime() - outage_data.index[0].to_pydatetime()
outage_data['Elapsed Days'] = outage_data['Elapsed DHMS'].apply(lambda x: float(x.total_seconds())/(3600*24)) 
```

```python
outage_data_parish = outage_data.groupby('Parish')
print 'Number of Parishes:', outage_data_parish.ngroups
#for name, group in outage_data_parish:
    #print name
    #print group
```

```python
figsize(8,6)
for name, group in outage_data_parish:
    plt.plot(group.index, group['Outage Percent'])
gca().set_xlabel('Date')
ax = gca()
ax.set_ylim(top=1.0)
gcf().autofmt_xdate()
```

```python
figsize(8,6)
for name, group in outage_data_parish:
    plt.plot(group['Elapsed Days'], group['Outage Percent'])
gca().set_xlabel('Elapsed Days')
ax = gca()
ax.set_ylim(top=1.0)
```
